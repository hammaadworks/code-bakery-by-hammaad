```python
def bool_recursive_target(index, sum, arr, target):
    if index == len(arr) and sum != target:
        print(index, sum, arr, target, "--> False")
        return False
    if sum == target:
        print(index, sum, arr, target, "--> True")
        return True
    
    print(index+1, sum+arr[index], arr, target)
    pick = bool_recursive_target(index+1, sum+arr[index], arr, target)
    print(index+1, sum, arr, target)
    not_pick = bool_recursive_target(index+1, sum, arr, target)
    
    return pick or not_pick

def bool_recursive_target_duplicate(index, sum, arr, target):
    if not index < len(arr) or sum > target:
        print(index, sum, arr, target, "--> False")
        return False
    if sum == target:
        print(index, sum, arr, target, "--> True")
        return True
    
    print(index+1, sum+arr[index], arr, target)
    pick_move = bool_recursive_target_duplicate(index+1, sum+arr[index], arr, target)
    print(index, sum+arr[index], arr, target)
    pick_hold = bool_recursive_target_duplicate(index, sum+arr[index], arr, target)
    print(index+1, sum, arr, target)
    not_pick = bool_recursive_target_duplicate(index+1, sum, arr, target)
    
    return pick_move or pick_hold or not_pick


def recursive_target(index, sum, arr, target, subset):
    if sum == target:
        return subset
    
    pass
    
def recursive_target_duplicate(index, sum, subset, arr, target):
    pass
    
arr = [1, 4]
target = 4
print(recursive_target(0, 0, [], arr,target))
print(bool_recursive_target(0, 0, arr, target))
print(recursive_target_duplicate(0, 0, [], arr, target))
print(bool_recursive_target_duplicate(0, 0, arr, target))
```
