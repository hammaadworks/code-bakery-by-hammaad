### Adjecency List

```python
class Graph:
    def __init__(self, nodes, edges):
        self.nodes = nodes
        self.adj_list = [[] for _ in range(nodes)]
        for n1,n2 in edges:
            ## Bidir Graph
            self.adj_list[n1].append(n2)
            self.adj_list[n2].append(n1)
```

```python
g = Graph(3,[(0,1),(0,2),(1,2)])
print(g.adj_list)
g
```
