
# WePay Transaction List REST-API - Mission 1.0  <hr>

Maintains a <u>super-simple list of transactions</u> done by customers on ***WePay***.

Scheme of our data-store looks like:

<h1>
    <table><th><td>Id</td><td>Customer Name</td><td>Transation Type</td><td>Transaction Amount</td><td>Bank</td></table>
</h1>

## Tech Stack <hr>

**Backend Framework:** Spring Boot Framework

**API Testing:** Postman

**Database:** In-memory H2 Database

**Misc:** Maven, Hibernate, Spring Initializr, Slf4j Logger
  
## API Reference<hr>

#### Get all transactions

```http
  GET /wepay
```

#### Get a transaction

```http
  GET /wepay/{Id}
```

#### Submit a transaction

```http
  POST /wepay
```

#### Update a transaction

```http
  PUT /wepay/{Id}
```

#### Delete a transaction

```http
  DELETE /wepay/{Id}
```
#### Get all transactions for a particular Bank

```http
  GET /wepay/bank/{Bank}
```
